# ABSA STOCK

```python
absa = data[data['Stock'] == ' ABSA Bank Kenya Plc Ord 0.50']
```

```python
#Calculate the log of the data in order to stabilise the variance of the time series data
absa['VWAP_log'].plot()
```

```python
#Stationarity
absa['Rolling_Mean'] = absa['VWAP'].rolling(window=3).mean()
plt.figure(figsize=(20,8))
plt.plot(absa['Date'], absa['VWAP'], label='Original Time Series')
plt.plot(absa['Date'], absa['Rolling_Mean'], label='Rolling Mean')
plt.legend()
plt.title('Original Time Series and Rolling Mean')
plt.show()
```

```python
#The p-values is greater than 0.05 hence the data is not stationary
result = adfuller(absa['VWAP_log'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])
```

```python
#Differencing
#This stabilises the mean of the data hence making the data stationary
absa['Diff']= absa['VWAP_log'].diff()
```

```python
absa = absa.dropna()
absa = absa.drop(columns='Stock')
absa = absa.set_index('Date')
```

```python
result = adfuller(absa['Diff'].dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
model = ARIMA(x['VWAP_log'], order=(2,1,2))
model_fit = model.fit()
print(model_fit.summary())
```

```python
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
residuals.plot(kind='kde')
print(residuals.describe())
```

```python
#Residuals acf and pacf have no significant spikes on the lags hence indicating presence of white noise
acf_res = plot_acf(residuals)
pacf_res = plot_pacf(residuals)
```

```python
def arima_train_and_plot(data, col_name, Ntest, p, d, q):
    # Define and fit the arima model
    model = ARIMA(x['VWAP_log'], order=(2,1,2))
    model_fit = model.fit()

    # Plot the real values of stock prices
    fig, ax = plt.subplots(figsize=(15,8))
    ax.plot(absa['VWAP_log'], label='Actual return')

    # Plot the fitted values of the model 
    train_pred = model_fit.fittedvalues
    ax.plot(x.index, train_pred, color='green', label='Fitted')

    # Plot the forecast values of the model 
    prediction_res = model_fit.get_forecast(Ntest)
    conf_int = prediction_res.conf_int()

    # Lower and upper limits of prediction
    lower, upper = conf_int[conf_int.columns[0]], conf_int[conf_int.columns[1]]
    forecast = prediction_res.predicted_mean
    ax.plot(y.index, forecast, label='Forecast')
    ax.fill_between(y.index, lower, upper, color='red', alpha=0.3)
    ax.legend()

    # Evaluating the model using RMSE and MAE metrics
    y_true = y['VWAP_log'].values
    rmse = np.sqrt(mean_squared_error(y_true, forecast))
    mae = mean_absolute_error(y_true, forecast)

    return rmse, mae
```

```python
#The model is relatively accurate
rmse, mae = arima_train_and_plot(absa, 'VWAP_log', Ntest=30, p=2, d=1, q=2)
print('Root Mean Squared Error: ', rmse)
print('Mean Absolute Error: ', mae)
```
